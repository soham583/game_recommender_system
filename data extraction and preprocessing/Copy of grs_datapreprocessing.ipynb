```python
import ast
```

```python
k = df['short_description'][0]
```

```python
def prostr(s):
  k = ''
  for i in s:
    if i == " ":
      k+=i
    elif i.isalnum():
      k+=i
    else:
      pass
  return k
```

```python
df['short_description'] = df['short_description'].apply(prostr)
```

```python
def cattag(s):
  s = ast.literal_eval(s)
  for i in range(len(s)):
    s[i] = "".join(c for c in s[i] if c.isalpha())
    s[i] = s[i].lower()
  return s
```

```python
df['categories'] = df['categories'].apply(cattag)
```

```python
df
```

```python
df['publisher'] = df['publisher'].apply(lambda x:["".join(c for c in x if c.isalpha())])
df['developer'] = df['developer'].apply(lambda x:["".join(c for c in x if c.isalpha())])
```

```python
df
```

```python
df['short_description'] = df['short_description'].apply(lambda x: list(x.split(" ")))
```

```python
df['genres'] = df['genres'].apply(lambda x:[x])
```

```python
df['tags'] = df['genres']+df['categories']+df['short_description']+df['developer']+df['publisher']
```

```python
df
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df1 = df[['steam_appid','name','tags']]
```
